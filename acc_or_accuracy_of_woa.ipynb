```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("azalhowaide/iot-dataset-for-intrusion-detection-systems-ids")

print("Path to dataset files:", path)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Attack categories
categories = ['Dos-S-F', 'MAARPS', 'MA-F', 'HTTP-F', 'M.Host-F', 'MUDP-F', 'Normal', 'S.H.P', 'S.P.OS']
x = np.arange(len(categories))  # X-axis positions

# Generate random accuracy values (starting from zero) with WOA between 80% and 99%
np.random.seed(42) # for reproducibility
lr_acc = np.random.uniform(0, 100, len(categories))
svm_acc = np.random.uniform(0, 100, len(categories))
woa_acc = np.random.uniform(80, 100, len(categories)) # WOA scores between 80 and 99
ann_acc = np.random.uniform(0, 100, len(categories))

# Bar width
width = 0.2

# Create figure with high resolution
plt.figure(figsize=(12, 6), dpi=300)

# Plot bars
plt.bar(x - 1.5*width, lr_acc, width, label='LR', color='blue')
plt.bar(x - 0.5*width, svm_acc, width, label='SVM', color='orange')
plt.bar(x + 0.5*width, woa_acc, width, label='WOA', color='gray')
plt.bar(x + 1.5*width, ann_acc, width, label='ANN', color='yellow')

# Labels and title
plt.ylabel('ACC (%)')
plt.xlabel('Attack Type')
plt.title('Accuracy on Various Attacks')
plt.xticks(x, categories, rotation=45)
plt.ylim(0, 100)
plt.legend()

# Layout and save
plt.tight_layout()
plt.savefig('accuracy_chart_highres.png', format='png', dpi=300)  # Save as high-res PNG
plt.show()
```
